```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df1 = pd.read_csv('Documentation/performance3_P2_1ms.csv',usecols = [' Context Switch Overhead',' Workload'])
#print(df1)
color=['r','b']

x1=df1[" Context Switch Overhead"]
y1=df1[" Workload"]
plt.figure(figsize=(10,10))
A=plt.scatter(x1,y1,s=5,c=color[0])
plt.xlabel(' Context Switch Overhead')
plt.yticks(fontsize = 'xx-small')
plt.ylabel('Workload (i*j*k)')
#plt.title(' Context Switch Overhead Time quanta 1ms')
#plt.show()


df2 = pd.read_csv('Documentation/performance3_P2_2ms.csv',usecols = [' Context Switch Overhead',' Workload'])
#print(df1)
x=df2[" Context Switch Overhead"]
y=df2[" Workload"]

B=plt.scatter(x,y,s=5)
plt.xlabel(' Context Switch Overhead')
plt.yticks(fontsize = 'xx-small')
plt.ylabel('Workload (i*j*k)')
plt.title(' Context Switch Overhead Time')
plt.legend((A,B),('quanta=1ms','quanta=2ms'),loc="lower right")

plt.show()
```
