```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df1 = pd.read_csv('Documentation/performance1.csv',usecols = ['Number of Threads',' Execution Time'],skiprows=0,nrows=1000)
df1.plot()
plt.xlabel('No. of Threads')
plt.xticks([0,100,200,300,400,500,600,700,800,900,1000])
plt.ylabel('Execution Time')
plt.title('workload=10*50 50*10')
plt.show()
```

```python
df2 = pd.read_csv('Documentation/performance1.csv',usecols = ['Number of Threads',' Execution Time'],skiprows=[i for i in range(1, 1001) ],nrows=1000)
df2.plot()
plt.xlabel('No. of Threads')
plt.xticks([0,100,200,300,400,500,600,700,800,900,1000])
plt.ylabel('Execution Time in ns')
plt.title('workload=100*500 500*100')
plt.show()

```
