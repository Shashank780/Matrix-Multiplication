```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df1 = pd.read_csv('Documentation/performance3_P2_1ms.csv',usecols = [' P2 Execution Time',' Workload'])
#print(df1)
color=['r','b']

x1=df1[" P2 Execution Time"]
y1=df1[" Workload"]
plt.figure(figsize=(10,10))
A=plt.scatter(x1,y1,s=5,c=color[0])
plt.xlabel('P2 Execution Time')
plt.yticks(fontsize = 'xx-small')
plt.ylabel('Workload (i*j*k)')
#plt.title('P2 Execution Time quanta 1ms')

df2 = pd.read_csv('Documentation/performance3_P2_2ms.csv',usecols = [' P2 Execution Time',' Workload'])
#print(df1)
x2=df2[" P2 Execution Time"]
y2=df2[" Workload"]

B=plt.scatter(x2,y2,s=5,c=color[1])
plt.xlabel('P2 Execution Time')
plt.yticks(fontsize = 'xx-small')
plt.ylabel('Workload (i*j*k)')
plt.title('P2 Execution Time ')
plt.legend((A,B),('quanta=1ms','quanta=2ms'),loc="lower right")

plt.show()
```
