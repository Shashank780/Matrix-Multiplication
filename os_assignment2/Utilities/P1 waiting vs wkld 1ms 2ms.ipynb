```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df1 = pd.read_csv('Documentation/performance3_P1_1ms.csv',usecols = [' P1 Waiting Time',' Workload'])
#print(df1)
color=['r','b']
x1=df1[" P1 Waiting Time"]
y1=df1[" Workload"]
plt.figure(figsize=(10,10))
A=plt.scatter(x1,y1,s=5,c=color[0])
plt.xlabel(' P1 Waiting Time')
plt.yticks(fontsize = 'xx-small')
plt.ylabel('Workload (i*j*k)')
#plt.title(' P1 Waiting Time quanta 1ms')


df2 = pd.read_csv('Documentation/performance3_P1_2ms.csv',usecols = [' P1 Waiting Time',' Workload'])
#print(df1)
x2=df2[" P1 Waiting Time"]
y2=df2[" Workload"]

B=plt.scatter(x2,y2,s=5,c=color[1])
plt.xlabel(' P1 Waiting Time')
plt.yticks(fontsize = 'xx-small')
plt.ylabel('Workload (i*j*k)')
plt.title(' P1 Waiting Time')
plt.legend((A,B),('quanta=1ms','quanta=2ms'),loc="lower right")

plt.show()
```
